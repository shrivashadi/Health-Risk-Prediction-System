```python
## Splitting dataset into x and y
y = df['Heart_Disease']
```

```python
x = df.iloc[:, :-1]
x.head()
```

```python
## Train test split
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

lr = LinearRegression()
lasso = Lasso()
ridge = Ridge()
```

```python
x_train.head()
```

```python
x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)
```

```python
x_train, x_test = x_train.align(x_test, join='left', axis=1, fill_value=0)
```

```python
lr.fit(x_train, y_train)
lasso.fit(x_train, y_train)
ridge.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
las_pred = lasso.predict(x_test)
ridge_pred = ridge.predict(x_test)
```

```python
## model evaluation
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print(y_test.shape)
print(lr_pred.shape)
```

```python
print(np.isnan(lr_pred).any())
print(np.isinf(lr_pred).any())
```

```python
y_test = y_test.ravel()
```

```python
print("y_test shape:", y_test.shape)
print("lr_pred shape:", lr_pred.shape)
print("las_pred shape:", las_pred.shape)
print("ridge_pred shape:", ridge_pred.shape)
```

```python
lr_rscore = r2_score(y_test, lr_pred)
las_rscore = r2_score(y_test, las_pred)
ridge_rscore = r2_score(y_test, ridge_pred)

print("Linear Regression R2:", lr_rscore)
print("Lasso R2:", las_rscore)
print("Ridge R2:", ridge_rscore)
```

```python
lr_pred = lr.predict(x_test)
las_pred = lasso.predict(x_test)
ridge_pred = ridge.predict(x_test)
```

```python
print(lr_pred.shape)
print(y_test.shape)
```

```python
lr_rscore = r2_score(y_test, lr_pred)
las_rscore = r2_score(y_test, las_pred)
ridge_rscore = r2_score(y_test, ridge_pred)

print("Linear Regression R2:", lr_rscore)
print("Lasso R2:", las_rscore)
print("Ridge R2:", ridge_rscore)
```

```python
lr_mse = mean_squared_error(y_test, lr_pred)
las_mse = mean_squared_error(y_test, las_pred)
ridge_mse = mean_squared_error(y_test, ridge_pred)

print("LR MSE:", lr_mse)
print("Lasso MSE:", las_mse)
print("Ridge MSE:", ridge_mse)
```

```python
print("Best Model: Ridge Regression")
print("Ridge R2 Score:", ridge_rscore)
print("Ridge MSE:", ridge_mse)
print("\nConclusion: Ridge Regression model performs the best")
print("with 99.1% accuracy on this dataset.")
print("It has the lowest MSE and highest R2 Score")
print("compared to Linear Regression and Lasso Regression.")
```
